## Cleaning Data
Now, the documented issues during assessment can be cleaned. 

**Note:** Making a copy of the original data before cleaning. Cleaning includes merging individual pieces of data according to the rules of [tidy data](https://cran.r-project.org/web/packages/tidyr/vignettes/tidy-data.html). The result should be a high-quality and tidy master pandas DataFrame.

```python
# Make copies of original pieces of data

tweet_archive_clean = we_rate_dogs.copy()
image_pred_clean = img_pred.copy()
tweet_json_clean = tweets.copy()
```

### Issue #1: Retweeted ratings and Replies in tweet archive

#### Define: Only original tweets are needed. Rows with retweeted_status_user_id and in_reply_to_user_id have to be removed from the dataset

#### Code

```python
# Remove retweets from tweet_archive_clean
tweet_archive_clean = tweet_archive_clean[pd.isnull(tweet_archive_clean['retweeted_status_user_id'])]
```

```python
# Remove retweets from tweet_archive_clean
tweet_archive_clean = tweet_archive_clean[pd.isnull(tweet_archive_clean['in_reply_to_user_id'])]
```

### Issue #2: Some Numerators are suspicious and some denominators not equal to 10

#### Define
The numerators and denominators were extracted based on the numbers before and after a "/" respectively.

Check the numerators and denominators with faulty values and correct them

#### Code

```python
# write a function that checks for names starting with lowercase,  
# and append them as "None" to an array. Then create a new column to dump the 
# list of new set names

dog_name = []
def wrong_name(dog):
    if dog['name'][0].islower() == True:
        dog_name.append('None')
    else:
        dog_name.append(dog['name'])
tweet_archive_clean.apply(wrong_name, axis=1)
tweet_archive_clean['dog_name'] = dog_name
```

```python
# drop the original name column

tweet_archive_clean.drop('name', axis=1, inplace=True)
```

#### Test

```python
# check for the counts of unique names in the name column

tweet_archive_clean.dog_name.value_counts()
```

### Insights:

## Which dog type is the favorite on Twitter?

```python
df.dog_stage.value_counts()
```

```python
dogs_df = df[df['dog_stage'] != 'None']
```

```python
dog_like = dogs_df.groupby('dog_stage', as_index=True).favorite_count.mean()
```

```python
dog_like
```

### Visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.barh(['doggo', 'floofer', 'pupper', 'puppo'], dog_like)
plt.xlabel('Average Likes')
plt.ylabel('Dog Stage')
plt.title('Average Likes by Dog Stage');
```

## Which dog stage has the average highest rating?

```python
dogs_df
```

```python
dog_ratings = dogs_df.groupby('dog_stage', as_index=True).rating_numerator.mean()
```

```python
dog_ratings
```

### Visualization

```python
plt.bar(['doggo', 'floofer', 'pupper', 'puppo'], dog_ratings)
plt.title('Dog stage by Average ratings on twitter',weight='bold', fontsize=16)
plt.xlabel('Dog Stage',weight='bold')
plt.ylabel('Average Ratings', weight='bold');
```

## Which dog stage is most tweeted about?

```python
dogs_df.groupby('dog_stage').tweet_id.count()
```

### Visualization

```python
plt.barh(['doggo', 'floofer', 'pupper', 'puppo'], 
       dogs_df.groupby('dog_stage').tweet_id.count())
plt.title('Dog Stage with most tweets')
plt.xlabel('Number of Tweets')
plt.ylabel('Dog Stage');
```
